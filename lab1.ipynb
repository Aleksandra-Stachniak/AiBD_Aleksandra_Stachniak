Poniżej znajduje się kod zdefiniowanej funkcji f(x)=x^2+5 i jej wykresu dla przedziału x>-1 oraz x<1.

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    y = x * x + 5
    return y

x1 = np.linspace(-1,1)
y = f(x1)

plt.plot(x1, f(x1))
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title("Funkcja f określona na przedziale <-1;1>")
plt.legend(["f(x) = x^2 + 5"])
plt.grid()
plt.show()
```

Poniżej znajduje się wykres funkcji f(x)=x^2+5 dla przedziału x>-6 oraz x<6.

```python
x2 = np.linspace(-6,6)
y = f(x2)

plt.plot(x2, f(x2))
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title("Funkcja f określona na przedziale <-6;6>")
plt.legend(["f(x) = x^2 + 5"])
plt.grid()
plt.show()
```

Poniżej znajduje się wykres funkcji f(x)=x^2+5 dla przedziału x>0 oraz x<5.

```python
x3 = np.linspace(0,5)
y = f(x3)

plt.plot(x3, f(x3))
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title("Funkcja f określona na przedziale <0;5>")
plt.legend(["f(x) = x^2 + 5"])
plt.grid()
plt.show()
```
